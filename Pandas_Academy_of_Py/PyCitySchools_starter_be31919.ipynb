```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

```python
#calculate % students with math score of 70% or greater
math_scores = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]

#calculate % students with reading score of 70% or greater
reading_scores = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
```

```python
#Total number of schools
total_schools = school_data_complete["School ID"].nunique()
print(total_schools)

#Total number of students
total_students = school_data_complete["Student ID"].count()
print(total_students)

#Total budget
total_budget = school_data["budget"].sum()
print(total_budget)
# still much convert budget to currency

#Average math score
avg_math_score = school_data_complete["math_score"].mean()
# avg_math_score_rnd = avg_math_score.round(6)
print(avg_math_score)

#Average reading score
avg_reading_score = school_data_complete["reading_score"].mean()
# avg_reading_score_rnd = avg_reading_score.round(6)
print(avg_reading_score)


#Percent passing Math
math_scores_perc = (len(math_scores)/total_students)*100
# math_scores_perc_rnd = math_scores_perc.round(6)
# print(math_scores_perc)
math_scores_perc_rnd = math_scores_perc.round(decimals=6)
print(math_scores_perc_rnd)

#Percent passing reading
reading_scores_perc = (len(reading_scores)/total_students)*100
# math_reading_perc_rnd = math_readings_perc.round(6)
# print(reading_scores_perc)
reading_scores_perc_rnd = reading_scores_perc.round(decimals=6)
print(reading_scores_perc_rnd)


# Overall passing rate
overall_passing_rate = (avg_math_score + avg_reading_score)/2
print(overall_passing_rate)
```

```python
district_summary = pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students], 
                                 "Total Budget":[total_budget],  "Average Math Score":[avg_math_score], 
                                 "Average Reading Score":[avg_reading_score], "% Passing Math":[math_scores_perc],
                                 "% Passing Reading":[reading_scores_perc], "% Overall Passing Rate":[overall_passing_rate]})
district_summary
```

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type

  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Create a new summary by schools with groupby and agg functions
grouped_school_summary = school_data_complete.groupby(["school_name"]).agg({
    "type": max, "Student ID": "count", "budget" : max, "math_score": "mean",
    "reading_score": "mean"})

#A dd column that calculates Per Student Budget
grouped_school_summary["Per Student Budget"] = grouped_school_summary["budget"] / grouped_school_summary ["Student ID"]

grouped_school_summary
```

```python
#c% students with math score of 70% or greater at each school
math_passing = school_data_complete[school_data_complete['math_score'] >= 70]
math_passing_summary = math_passing.groupby(['school_name']).agg({'Student ID': 'count'})
math_passing_summary.columns = ['Count of Students Passing Math']
grouped_school_summary['% Passing Math'] = (math_passing_summary['Count of Students Passing Math']
                                            /grouped_school_summary['Student ID'])*100

#% students with reading score of 70% or greater at each school
reading_passing =school_data_complete[school_data_complete['reading_score'] >= 70]
reading_passing_summary = reading_passing.groupby(['school_name']).agg({'Student ID': 'count'})
reading_passing_summary.columns = ['Count of Students Passing Math']
grouped_school_summary['% Passing Reading'] = (reading_passing_summary['Count of Students Passing Math']
                                               /grouped_school_summary['Student ID'])*100

#students overall passing rate at each school
grouped_school_summary["% Overall Passing Rate"] = (grouped_school_summary["% Passing Math"] +
                                                    grouped_school_summary["% Passing Reading"])/2

grouped_school_summary
```

```python
#Format table columns
grouped_school_summary.rename(columns = {"type": "School Type", "Student ID" : "Total Students", 
                                         "budget": "Total School Budget","math_score": "Average Math Score", 
                                         "reading_score":"Average Reading Score"}, inplace=True)
# grouped_school_summary

#Correct column locations to match assignment
grouped_school_summary = grouped_school_summary[["School Type","Total Students", "Total School Budget", 
                                                 "Per Student Budget","Average Math Score","Average Reading Score", 
                                                 "% Passing Math", "% Passing Reading", "% Overall Passing Rate" ]]


grouped_school_summary
```

```python
grouped_school_summary_formatted = grouped_school_summary.copy()
grouped_school_summary_formatted
```
